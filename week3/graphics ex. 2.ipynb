 In PHYS 212, many of you encountered the wave function shown in question 1(a)
(Equation (1)). You plotted this along with the potential energy from the positive energy
solutions for Schroedinger’s equation:

$
U (x) = −\dfrac{1}{\cosh^2(ax)}
$

Reproduce the following graph of these two functions for the four values of k shown. Let
A and a = 1.



```python
from numpy import *
import matplotlib.pyplot as plt

x = arange(-pi, pi, .1)
kvals = [0.0, .5, 1.0, 2.0]
i = 0 + 1j
Upot = -1/ cosh(x)

plot(x, Upot, label = '$U(x)$', lw = '2')
xlabel('$x (rad)$', fontsize = 14)
ylabel('$U(x), \psi^2(x)$', fontsize = 14)


k = 0
psi = ((i*k - tanh(x))/(i*k +1))*e **(i*k*x)
psi2 = psi * psi.conjugate()
plot(x, psi2, label = '$\Psi^2, k=0$', color = 'red', lw = '2')

k = .5
psi = ((i*k - tanh(x))/(i*k +1))*e **(i*k*x)
psi2 = psi * psi.conjugate()
plot(x, psi2, label = '$\Psi^2, k=.5$', color = 'red', lw = '2', ls ='--')

k = 1
psi = ((i*k - tanh(x))/(i*k +1))*e **(i*k*x)
psi2 = psi * psi.conjugate()
plot(x, psi2, label = '$\Psi^2, k=1$', color = 'red', ls = '-.')

k = 2
psi = ((i*k - tanh(x))/(i*k +1))*e **(i*k*x)
psi2 = psi * psi.conjugate()
plot(x, psi2, label = '$\Psi^2, k=2$', color = 'red', lw = '2', ls =':')

xticks(arange(-pi, pi, pi/2), ['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi/2$', '$\pi$'])
legend(loc=4)
show()
```

On PolyLearn, you will find a file called sunspots.txt, (See the note at the bottom
of this file for saving this file from PolyLearn in the proper format.) which contains the
observed number of sunspots on the Sun for each month since January 1749. The file contains two columns of numbers, the first being the month and the second being the sunspot
number.

(a)
-
Write some code that reads in the data and makes a graph of sunspots as a function of time. Format the x-axis labels to show the year. You will have to look up documentation for how to do this. Be sure to think about how to make this graph publication quality, i.e. it should have labels, a legend, and customized (i.e. NOT default) color and data representation choices. The following can be used to extract the data from the file into numpy arrays:


```python
data = loadtxt("/home/ssteeper/Desktop/sunspots.txt",float)
x = data[:,0]
y = data[:,1]
plot(x, y, lw=.5, c='red', label='sunspots')
xlabel('Year', fontsize = 14)
ylabel('Sunspots', fontsize = 14)
title('Sunspots vs time since 1749', fontsize =14)
plt.xticks(arange(0, 3000, 480), ['1749', '1789', '1829', '1869', '1909', '1949', '1989'])
legend()
show()
```

(b)
-
Now write code to display only the first 1000 data points on the graph.


```python
print len(x), type(x)
```

```python
a = np.zeros(1000)
b = np.zeros(1000)
for i in range(1000):
    a[i] = x[i]
    b[i] = y[i]
    
plot(a, b, lw=.5, c='red', label='sunspots')
xlabel('Year', fontsize = 14)
ylabel('Sunspots', fontsize = 14)

title('Sunspots vs time since 1749, first 1000 months', fontsize =14)
plt.xticks(arange(0, 1200, 480), ['1749', '1789', '1829'])
legend()
show()
```

(c)
-

Finally, add code to calculate the running average of the data. Plot both the original
data and the running average on the same graph, again over the range covered by the
first 1000 data points. Be sure to include a legend and make it “pretty” (shorthand for
publication quality).


```python
a = np.zeros(1000)
b = np.zeros(1000)
avg = np.zeros(1000)
tot = 0
for i in range(1000):
    a[i] = x[i]
    b[i] = y[i]
    tot += y[i]
    avg[i] = tot / i

    
plot(a, b, lw=.5, c='red', label='sunspots')
plot(a, avg, lw=2, c='blue', label='running avg')
xlabel('Year', fontsize = 14)
ylabel('Sunspots', fontsize = 14)
title('Sunspots vs time since 1749, first 1000 months', fontsize =14)
plt.xticks(arange(0, 1200, 480), ['1749', '1789', '1829'])
legend()
show()
```

There is another file on PolyLearn called stars.txt, (See the note at the bottom of this
file for saving this file from PolyLearn in the proper format.) which contains the temper-
atures and magnitudes of several thousand stars in the Milky Way near our solar system.
You will use this data to construct a scatter plot of stellar magnitude vs. temperature,
called a Hertzsprung-Russell diagram, after the astronomers who first plotted such data
this way. The diagram is one of the fundamental tools of stellar astrophysics. For historical reasons, H-R diagrams are normally plotted with both the magnitude and temperature
axes decreasing rather than increasing.



(a)
-
Read in the data from stars.txt using similar commands as in question 3. Plot the
data as is, with magnitude and temperature increasing in the usual way. Be sure to properly format your graph as publication quality.


```python
data = loadtxt("/home/ssteeper/Desktop/stars.txt",float)
x = data[:,0]
y = data[:,1]
plot(x, y, '.', c='green')
#grid(True)
title('Hertzsprung-Russell Diagram')
xlabel('Temperature')
ylabel('Magnitude')
show()
```

(b)
-
Now use the xlim() and ylim() commands to flip your graph so that it is a correct
H-R diagram. There are several regions of the diagram where stars seem to cluster. Do
a web search on H-R diagrams to find out what each of these regions are called and add
some annotations to your graph to identify them.


```python
plot(x, y, '.', c='yellow')
#grid(True)
title('Hertzsprung-Russell Diagram')
xlabel('Temperature')
ylabel('Magnitude')
xlim(16000,0)
ylim(20, -5)
annotate('White Dwarfs', (9000,17), xytext=None)
annotate('Giants', (3800,2), xytext=None)
annotate('Main Sequence', (3600,6.5), xytext=None)
show()
```
