Using matplotlib to make pretty graphs
-

```python
#simple line plot
t = arange(0.0,1.01,.01)
c = cos(pi * 4 * t)
s = sin(pi * 4 * t)
```

```python
plot(t, s, linewidth = 2, label='$sin(4{\pi}t)$')    #note markdown syntax for labels
plot(t, c, color="red", linewidth = 2, label='$cos(4{\pi}t)$') 

xlabel('time (seconds)', fontsize = 14)
ylabel('Voltage (mV)', fontsize = 14)
title('Voltage vs time graph', fontsize = 20)


legend()#for the line labels
savefig("simple_plot.pdf")
show()
```
