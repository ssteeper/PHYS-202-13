```python
x = arange(-5, 5, .1)
f = (x+1)/(x-1)
g = e**(-x)

plot(x,f, c='b', lw=2, label = '(x+1)/(x-1)')
plot(x,g, ':', c='r', lw=3, label = 'exp(-t)')  
xlim((-5,5))
ylim(-10,10)

legend(loc = 'lower left')
xlabel('time(seconds)', fontsize = 16)
ylabel('f(t), g(t)', fontsize =16)
annotate('discontinuous', (1.3, -2.5))
annotate('here', (1.3, -3.5))

show()
```

```python
from Electrostatics import *   #copied Electrostatics.py into Midterm directory
from scipy import integrate

a = 1.0   #m
b = 2.0   #m
qd = 1e-6  #Charge Density in C/m
dx = .1  #m
q = qd*dx

Xq = arange(a,b,dx)
Yq = np.zeros(len(Xq))
x0 = 3.0  #m
y0 = 3.0  #m

Ex = pointField(x0,y0,q,Xq,Yq)[0]
Ey = pointField(x0,y0,q,Xq,Yq)[1]
Ix = integrate.trapz(Ex)
Iy = integrate.trapz(Ey)
print 'Ex: ' + str(Ix) + ' N/C'
print 'Ey: ' + str(Iy) + ' N/C'
```

```python
qd0 = 2e-6   #C/m
a, b = 1.0, 3.0   #m
x0, y0 = 5.0, 3.0  #m
dx = .1  #m
qd = lambda x: qd0*(x**2)/(b**2)  

Xq = arange(a,b,dx)
Yq = np.zeros(len(Xq))
q = qd(Xq)*dx

Ex = pointField(x0,y0,q,Xq,Yq)[0]
Ey = pointField(x0,y0,q,Xq,Yq)[1]
Ix = integrate.trapz(Ex)
Iy = integrate.trapz(Ey)

print 'Ex: ' + str(Ix) + ' N/C'
print 'Ey: ' + str(Iy) + ' N/C'
```
