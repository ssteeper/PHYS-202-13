Sampling from a continuous distribution
-

```python
x = arange(0., pi+0.01, .01)
f = lambda x: sin(x)/2
F = lambda x: (cos(0) - cos(x))/2

plot(x, f(x), label = 'True', lw = 3)
plot(x, F(x), label = 'Cumulative function', lw = 3)

xlim(0, 3.5)
ylim(0, 1.1)

legend(loc=0)
show()
```

```python
N = 10000
z = numpy.random.random_sample(N)
Finv = lambda z: arccos(cos(0) - 2*z)
xsamples = Finv(z)
hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f(x),label='True', lw=3, c='r')

xlim(0, pi)
ylim(0, .7)
legend(loc=0)
show()
```

Exercise 1. Exponential function - analytic

Determine the CDF $F(x)$ for this function analytically and invert it analytically to obtain $F^{-1}$.

```python
L=1
x = arange(0., 5., .01)
f = lambda x: L*e**(-L*x)
F = lambda x: -e**(-L*x)
Finv = lambda x: - log(x)/L
```

(b)

Plot the function and CDF on the same plot over the range [0,5]. Let $\lambda = 1$.

```python
plot(x, f(x), lw=2, label='$f(x)$')
plot(x, F(x), lw=2, label='$F(x)$')
xlabel('x', fontsize = 18)
ylabel('y', fontsize = 18)
legend()
show()
```

Exercise 2. Exponential function - numerical
-

(a)
-
Write a program to implement the numerical inverse CDF transform algorithm for the exponential function from Exercise 2. You may wish to consult the numpy documentation for using the functions "cumsum" and "searchsorted". Let $\lambda = 1$.

```python
# I.
L=1
x = arange(0., 5., .01)
distrib = L*e**(-L*x)
```

```python
# II.
CDF = np.zeros(len(distrib))
total = sum(distrib)
for i in range(len(distrib)):
    CDF[i] = cumsum(distrib)[i]/total
```

```python
# III.
N = 10000
z = numpy.random.random_sample(N)
```

```python
# IV, V
y = np.zeros(len(distrib))
for i in z:
    ndx = CDF.searchsorted(i)
    y[ndx] = i
```
