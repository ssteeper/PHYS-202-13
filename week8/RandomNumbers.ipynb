Exercise 2. Exponential function - numerical
-

(a)
-
Write a program to implement the numerical inverse CDF transform algorithm for the exponential function from Exercise 2. You may wish to consult the numpy documentation for using the functions "cumsum" and "searchsorted". Let $\lambda = 1$.

```python
# I.
L=1
x = arange(0., 5., .01)
distrib = L*e**(-L*x)
```

```python
# II.
CDF = np.zeros(len(distrib))
total = sum(distrib)
for i in range(len(distrib)):
    CDF[i] = cumsum(distrib)[i]/total
```

```python
# III.
N = 10000
z = numpy.random.random_sample(N)
```

```python
# IV, V
xsamples = np.zeros(len(distrib))
for i in z:
    ndx = CDF.searchsorted(i)
    xsamples[ndx] = i
```

(b)
-
Make a plot of the true distribution, the numerically determined CDF, and a histogram of 10,000 sampled values using 50 bins, on the same graph.

```python
plot(x, distrib, c='r', lw=2, label = 'distrib')
plot(x, CDF, c='g', lw = 2, label = 'CDF')
legend()
show()
hist(xsamples, bins = 50)
xlim(0,5)
show()
```

(c)
-

Turn your algorithm from (a) into a reusable function that can be used for any 1-dimensional numerical distribution.

```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data."""
    Fx = cumsum(prob)/sum(prob)
    z = numpy.random.random_sample(N)
    xsamples = x[searchsorted(Fx,z)]
    return xsamples
```

```python
x = arange(0., 5., .01)
prob = x
N = 10000

xsamples = distribute1D(x, prob, N)
```

```python
hist(xsamples, bins = 50)
show()
```

Exercise 3. Quantum double-slit experiment
=


```python
d = 15e-6
a = 10e-6
L = 1
wl = 632.8e-9
screen = .2*.2

x = arange(-.2, .2, .001)
N = 10000

ISS = 2*sinc(a*x/(wl*L))**2
IDS = cos(2*pi*d*x/(wl*L))**2
prob = ISS * IDS
```

(b)
-
Make a 1D plot of the true distribution with the sampled distribution as a histogram to verify that it does indeed accurately simulate the original intensity distribution.

```python
xsamples = distribute1D(x, prob, N)
plot(x, prob)
show()
hist(xsamples, bins=100)
show()
```

```python
curve = lambda x: (sinc(a*x/(wl*L))**2) * (cos(2*pi*d*x/(wl*L))**2)
N =1

distrib = distribute1D(x,curve(x),N)
n,bins,patches=hist(distrib,500,normed=True,label="1 photon")
plot(x,max(n)*curve(x))
xlabel("x (m)", fontsize = 16)
ylabel("intensity", fontsize = 16)
legend()
show()
```
