CDF of Weibull distribution:
\begin{equation}
F(x) = 1 - e^{-(x/\beta)^{\alpha}}
\end{equation}

(b)

Inverse of CDF:
\begin{equation}
F^{-1}(x)=\left(-\beta^{\alpha}\ln(1-x)\right)^{1/\alpha}
\end{equation}

(c)

Write a program to generate 10,000 samples from the Weibull distribution

```python
N = 10000
z = numpy.random.random_sample(N)

a=1.5
b=6.0

f = lambda x: a * b**(-a) * x**(a-1) * exp(-(x/b)**a)
F = lambda x: 1 - exp((-(x/b)**a))
Finv = lambda x: (-b**a * log(1-x))**(1./a)

xsamples = Finv(z)
```

(d)
-
Plot the true distribution, CDF, and sampled distrib on the same graph.

```python
x = arange(0.,35.01,0.01)
hist(xsamples, 50, normed=True, label="Sampled Distribution", color = 'g')
plot(x, F(x), 'k', lw=2, label="CDF")
plot(x, f(x), 'r-', lw=1.5, label="True Distribution")
xlim(0,20)
legend()
show()
```
