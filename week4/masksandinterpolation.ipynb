Using Boolean Masks for Finding Data
==

```python
theta = arange(0., 2*pi, pi/100)
plot(theta, sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
```

```python
y = sin(theta[mask])
jj = (y==max(y))
y[jj] = NaN

plot(theta, sin(theta))
plot(theta[mask], y, lw=2, c='red')
show()
```

```python
theta = arange(0.,2*pi+.01,pi/10)
mask = (theta == .5)
print mask
```

```python
newTheta = .5; newSin = 0.
```

```python
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp = interp1d(theta,sin(theta))
newSin = sinInterp(newTheta)
```

```python
plot(theta, sin(theta), newTheta, newSin, 'r+', markersize=20, markeredgewidth=2)
show()
```

```python
thetaNew = arange(0,2*pi+0.1, pi/30)
sinNew = interp1d(theta, sin(theta))(thetaNew)
plot(theta, sin(theta), thetaNew, sinNew, 'r.')
show()
```

```python
plot(thetaNew, sinNew-sin(thetaNew))   #the difference between the interpolation and actual curve
show()
```

```python
sinNewC = interp1d(theta, sin(theta), kind='cubic')(thetaNew)
subplot(211)
plot(thetaNew, sinNew-sin(thetaNew), 'b', thetaNew, sinNewC - sin(thetaNew), 'r')
subplot(212)
plot(thetaNew, sinNewC - sin(thetaNew), 'r')
show()
```

Numerical Differentiation
=

```python
x = arange(0., 10.1, .1)
h = .1
y = x**5
dydx = ((x+h)**5 + x**5)/h
subplot(211)
plot(x, y)
subplot(212)
plot(x, dydx)
show()
```

```python
dy = diff(y)  #y(n+1) - y(n)
dx = diff(x)
dydx = dy/dx
```

```python
dydx1 = zeros(y.shape,float)
```

```python
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
```

```python
dydxExact = 5*x**4
plot(x, dydx1, x, dydxExact)
show()
```
