Electric Dipole Potential
-

$
V(x,y) = \dfrac{kq}{\sqrt{x^2+(y-d)^2}}-\dfrac{kq}{\sqrt{x^2+(y+d)^2}}
$

```python
dx, dy = .06,.06
x = arange(-2., 2., dx)
y = arange(-2., 2., dy)
#print x.ndim
#print len(x)

X,Y = meshgrid(x,y)
#print X.ndim
#print Y.shape

k=1.; q=1.; d=1.
V=(k*q)/(((X**2)+(Y-d)**2))**(1/2.)-(k*q)/((X**2)+(Y+d)**2)**(1/2.)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, lw=0, cmap=cm.prism)
cb = fig.colorbar(p, shrink=.5)
```

```python
#Represent 3D data by contours
#levels = array([8, 4, 2, 1, .5, .25, 0, -.25, -.5, -1, -2, -4, -8])
levels = linspace(8.03, -8.012, num=100)

contour(X,Y,V,levels,cmap=cm.flag)
show()
```

The first function should be named pointPotential(). It should take as arguments the point or points in space where you want to know the potential, the size (and sign) of the charge q and the position of the charge in the plane, and return the electric potential due to that charge.

```python
def pointPotential(x, y, q, posx, posy):
    """Computes electric potential from charge q at posx, posy and returns the potential at x,y"""
    k = 8.98755178e9 
    return (k*q)/sqrt((x-posx)**2 + (y-posy)**2)
```

Create another function dipolePotential(), that takes as arguments the point or points in space where you want to know the potential, the size of the charges q, and the separation distance d between the charges. This function should call the pointPotential function you created in part(a) and return the electric potential due to the dipole. For simplicity, let the two charges lie along the x-axis with their midpoint at the origin. 

```python
def dipolePotential(x,y,q,d):
    return pointPotential(x,y,q,-d,0) + pointPotential(x,y,q,d,0)
```

```python
x,y, = meshgrid(arange(-4.,4.,.06),arange(-4.,4.,.06))
q = 1e-9
d = 2

V = dipolePotential(x,y,q,d)

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, V, rstride=1, cstride=1, lw=0, cmap=cm.prism)
cb = fig.colorbar(p, shrink=.5)
```

```python
def quadrupolePotential(x,y,q,d):
    return pointPotential(x,y,q,-d,0) + pointPotential(x,y,q,d,0) + pointPotential(x,y,q,0,-d) + pointPotential(x,y,q,0,d)

V = quadrupolePotential(x,y,q,d)

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, V, rstride=1, cstride=1, lw=0, cmap=cm.prism)
cb = fig.colorbar(p, shrink=.5)
```
