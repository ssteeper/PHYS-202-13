3D Plotting and Functions
=


Electric Dipole Potential
-

$
V(x,y) = \dfrac{kq}{\sqrt{x^2+(y-d)^2}}-\dfrac{kq}{\sqrt{x^2+(y+d)^2}}
$

```python
dx, dy = .06,.06
x = arange(-2., 2., dx)
y = arange(-2., 2., dy)
#print x.ndim
#print len(x)

X,Y = meshgrid(x,y)
#print X.ndim
#print Y.shape

k=1.; q=1.; d=1.
V=(k*q)/(((X**2)+(Y-d)**2))**(1/2.)-(k*q)/((X**2)+(Y+d)**2)**(1/2.)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, lw=0, cmap=cm.prism)
cb = fig.colorbar(p, shrink=.5)
```

```python
#Represent 3D data by contours
#levels = array([8, 4, 2, 1, .5, .25, 0, -.25, -.5, -1, -2, -4, -8])
levels = linspace(8.03, -8.012, num=10000)

contour(X,Y,V,levels,cmap=cm.flag)
show()
```

Make a 3-D plot of the function

$
G(x,y) = e^{(-x^2-y^2)}
$

over the range -6 < $(x,y)$ < 6 using 51 bins in $x$ and 51 bins in $y$.

```python
dx, dy = 12/51.0,12/51.0
xmin, xmax = -6., 6.
ymin, ymax = -6., 6.
x = arange(xmin, xmax, dx)
y = arange(ymin, ymax, dy)
nbinsx = (2*xmax)/dx
nbinsy = (2*ymax)/dy

X,Y = meshgrid(x,y)
G = e**(-X**2-Y**2)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, G, rstride=1, cstride=1, lw=0, cmap=cm.prism)
cb = fig.colorbar(p, shrink=.5)
```

```python
levels = linspace(-3.01, 3.01012, num=100)
contour(X,Y,G,levels,cmap=cm.flag)
show()
```

```python
k = 8.98755178e9  #N * m/C**2
q = 10**(-9)      #coulombs
d = 2             #meters
x = arange(-4., 4., dx)
y = arange(-4., 4., dy)
X,Y = meshgrid(x,y)
V = (k*q)/sqrt((X**2)+(Y-d)**2) + (k*q)/sqrt((X**2)+(Y+d)**2) - (k*q)/sqrt((X-d)**2+(Y**2)) - (k*q)/sqrt(((X+d)**2)+(Y**2))

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, lw=0, cmap=cm.prism)
cb = fig.colorbar(p, shrink=.5)
```

```python
levels = linspace(-2.1, 2.3, num=1000)
contour(X,Y,V,levels,cmap=cm.flag)
show()
```
