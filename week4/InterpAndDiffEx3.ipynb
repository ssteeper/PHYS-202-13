```python
#from Electrostatics import *
#the import was not working correctly. after copy-pasting the definitions in that file to this notebook, it worked fine.
def pointPotential(x, y, q, posx, posy):
    """Computes electric potential from charge q at posx, posy and returns the potential at x,y"""
    k = 8.98755178e9 
    return (k*q)/((x-posx)**2 + (y-posy)**2)**(.5)

def quadrupolePotential(x,y,q,d):
    """Computes potential at x,y in a quadrupole system with charges at +d and -d along the x and y axis"""
    return pointPotential(x,y,q,-d,0) + pointPotential(x,y,q,d,0) + pointPotential(x,y,q,0,-d) + pointPotential(x,y,q,0,d)

def dipolePotential(x,y,q,d):
    """Computes potential at x,y in a dipole system with charges at +d and -d along the x axis"""
    return pointPotential(x,y,q,0,-d) - pointPotential(x,y,q,0,d) 

x,y = meshgrid(arange(-5,5,0.5),arange(-5,5,0.5))
Vdipole = dipolePotential(x,y,1e-9,2.)
contour(x,y,Vdipole,cmap=cm.prism)
show()
```

arrays.

```python
def finiteDifference(x,y):
    """takes two arrays, x and y, and returns the derivative of y with respect to x, dydx"""
    dydx1 = zeros(len(y),float)
    dydx1[:-1] = diff(y)/diff(x)
    dydx1[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    return dydx1
```

(b)
-
Use your function to take the derivative of the function $y = x^5$ over the range $0 < x < 10$ , with steps of .1.

```python
x=arange(0,10,.1)
y=x**5
dydx = finiteDifference(x,y)
subplot(211)
plot(x,dydx)
title('using finite difference')
xticks([])
subplot(212)
title('true derivative')
plot(x,5*x**4)
show()
```

I know my derivative function works because it matches the true derivative, $5x^4$ , exactly.

(c)

What is the accuracy of the result if you use a spacing of 0.001? (Plot the result)

```python
x=arange(0,10,.001)
y=x**5
dydx = finiteDifference(x,y)
subplot(211)
plot(x,dydx)
title('using finite difference')
xticks([])
subplot(212)
title('difference between true and numerical')
plot(x,5*x**4 - dydx)
show()
```

```python
def fourPtFiniteDiff(x,y):
    """Returns derivative dydx using four point diff method"""
    dydx = zeros(len(y), float)
    dydx[2:-2] = (y[0:-4]-(8*y[1:-3]) + (8*y[3:-1])-y[4:])/(12*(x[3]-x[2]))
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```
