```python
from mpl_toolkits.mplot3d import Axes3D

dx,dy = .03,.03
x = arange(-1., 1., dx)
y = arange(-1., 1., dy)
X,Y = meshgrid(x,y)

V = pointPotential(X,Y,(-2e-9),0,-.5) + pointPotential(X,Y,(3e-9),.5,0) + pointPotential(X,Y,(-3e-9),-.5,0)

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, lw=0, cmap=cm.prism)
cb = fig.colorbar(p, shrink=.5)
```
