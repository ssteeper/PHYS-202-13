```python
def deriv(theta_current, t):
    g = 9.8; l = .5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([dtheta, domega])
```

```python
from scipy.integrate import odeint
t = arange(0.,10.+.01,.01)

omega0 = 0.
th0 = arange(0, pi+.1, pi/4)
```

```python
theta = np.zeros(len(th0))
labels = ['0', 'pi/4', 'pi/2', '3pi/4', 'pi']
i = 0
for theta in th0:
    theta_init = array([theta, omega0])
    theta = odeint(deriv, theta_init, t)
    plot(t, theta[:,0], lw=1, label = labels[i])
    i+=1
legend()
show()
```

As the initial angle of the pendulum increases from $\pi/4$ to $3\pi/4$, the period increases.
