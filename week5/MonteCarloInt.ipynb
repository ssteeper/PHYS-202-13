Repeat the hit or miss method to .001% accuracy.

```python
f = lambda x: x**(.5) + sin(x)
x = arange(0.,15.,.001)

xmin = 1.7
xmax = 12.4

subx = x[logical_and((x > xmin), (x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot([xmin,xmin],[ymin,ymax],c='k', ls='--')
plot([xmax,xmax],[ymin,ymax],c='k', ls='--')
plot([xmin,xmax],[ymax,ymax],c='k', ls='--')
xlabel('$x$', fontsize=16)
ylabel('$x^{1/2} + sin(x)$', fontsize=16)
show()
```

```python
N = 1000000
samples_x = xmin + (xmax-xmin)*random_sample(N)
samples_y = ymin + (ymax-ymin)*random_sample(N)

myplot = plot(x,f(x),lw=3)
xlim(0.,15.)
plot(samples_x[1:200],samples_y[1:200],'.')

plot([xmin,xmin],[ymin,ymax],c='k', ls='--')
plot([xmax,xmax],[ymin,ymax],c='k', ls='--')
plot([xmin,xmax],[ymax,ymax],c='k', ls='--')
```

```python
newmask = (samples_y < f(samples_x))
sum(newmask)
```

Analytic Integral: $\frac{2}{3}x^{3/2} - cos(x)$

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax)-Iana(xmin)
IerrMC = 100*(Iexact-I)/Iexact
print "I_exact =",Iexact
print "I_hitormiss =", I
print "error on I =", abs(IerrMC),"%"
```

(a) Write code to compute the integral of the function $f(x) = x^{1/2} + sin(x)$ from example 1 using the sample mean method for $N=10000$   .

```python
f = lambda x: x**(.5) + sin(x)
N=10000
tot=0.0
samples_x = xmin + (xmax-xmin)*random_sample(N)

k = (xmax-xmin)/N
for i in range(N):
    tot += f(samples_x[i])
I = tot*k
Ierr = 100*(Iexact-I)/Iexact
print "I_sampleMean =", I
```

(b)
Plot the function and the average value of the function for these samples on the same graph. The area under the average is approximately equal to the area under the curve.


```python
avg = tot/N
myplot = plot(x,f(x),lw=2, label='$f(x)$')
xlim(0.,15.)

plot([xmin,xmin],[ymin,ymax],c='k', ls='--')
plot([xmax,xmax],[ymin,ymax],c='k', ls='--')
plot([0,15],[avg,avg], ls='-', c='red', label='average')

xlabel('$x$', fontsize=16)
ylabel('$x^{1/2} + sin(x)$', fontsize=16)
legend()
show()

```

(c) Compute the percent difference between the integral using the sample mean method and the exact value. How does the sample mean method result compare to the hit or miss method?

```python
f = lambda x: x**(.5) + sin(x)
N=1000000
tot=0.0
samples_x = xmin + (xmax-xmin)*random_sample(N)

k = (xmax-xmin)/N
for i in range(N):
    tot += f(samples_x[i])
I = tot*k
Ierr = 100*(Iexact-I)/Iexact
print "I_exact =",Iexact
print "I_sampleMean =", I
print "sampleMean error on I =", abs(Ierr),"%"
print "monteCarlo error on I =", abs(IerrMC),"%"
```

The sample mean method result is less accurate than the hit or miss method for the same number of samples, but not by much.
